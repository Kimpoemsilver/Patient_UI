```python
def show_tables(conn):
    cur = conn.cursor()
    cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tables = [r[0] for r in cur.fetchall()]
    return tables

conn = sqlite3.connect(DB_PATH)
tables = show_tables(conn)
print("table list:", tables)
conn.close()
```

```python
def preview_table(db_path, table, limit=10):
    conn = sqlite3.connect(db_path)
    cur = conn.cursor()
    cur.execute(f"SELECT * FROM {table} LIMIT {limit}")
    rows = cur.fetchall()
    col_names = [desc[0] for desc in cur.description]
    conn.close()
    print(f"\n {table} (ìƒìœ„ {limit}ê°œ):")
    print(col_names)
    for r in rows:
        print(r)

# ì˜ˆì‹œ ì‹¤í–‰
preview_table(DB_PATH, "patient_daily", limit= 1)
```

```python
import sqlite3
from datetime import date, timedelta
import random

DB_PATH = "project_db2.db"

patients = [
    {"id":"test001","pw":"pw001","name":"í™ê¸¸ë™","sex":1,"age":36,"height":170,"weight":65,
     "egfr":92,"HSI":5,"phq9":12,"birth_date":"1988-05-10","first_visit_date":"2025-09-10"},
    {"id":"test002","pw":"pw002","name":"ì´ëª½ë£¡","sex":1,"age":29,"height":180,"weight":75,
     "egfr":100,"HSI":3,"phq9":10,"birth_date":"1995-09-22","first_visit_date":"2025-09-10"}
]

with sqlite3.connect(DB_PATH) as conn:
    cur = conn.cursor()

    for p in patients:
        # ------------------------
        # 0. í™˜ì ê¸°ë³¸ì •ë³´ ë“±ë¡
        # ------------------------
        cur.execute("INSERT OR REPLACE INTO patient (patient_id, pw) VALUES (?, ?)", (p["id"], p["pw"]))
        cur.execute("""
            INSERT OR REPLACE INTO patient_info
            (patient_id, sex, age, height, weight, egfr, HSI, phq9, name, birth_date, first_visit_date)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (p["id"], p["sex"], p["age"], p["height"], p["weight"],
              p["egfr"], p["HSI"], p["phq9"], p["name"], p["birth_date"], p["first_visit_date"]))

        # ------------------------
        # 1. ì²« ì²˜ë°© (2025-09-10)
        # ------------------------
        presc_date1 = "2025-09-10"
        dose1 = 20 if p["id"]=="test001" else 10
        freq1 = 24 if p["id"]=="test001" else 12
        days1 = 7
        day_drug1 = int(24/freq1)
        cur.execute("""
            INSERT OR REPLACE INTO patient_predict
            (patient_id, prescription_date, dose, frequency, prescription_days, day_drug, note)
            VALUES (?, ?, ?, ?, ?, ?, ?)
        """, (p["id"], presc_date1, dose1, freq1, days1, day_drug1, "ì²« ë°©ë¬¸ ì²˜ë°©"))

        # PK param ì´ˆê¸°ê°’
        CL1 = round(15 + random.uniform(-2, 2), 2)
        V1 = round(120 + random.uniform(-10, 10), 2)
        cov = f'{{"WT":{p["weight"]},"eGFR":{p["egfr"]}}}'
        cur.execute("""
            INSERT OR REPLACE INTO pk_param (patient_id, pkpram_CL, pkpram_V, covariate)
            VALUES (?, ?, ?, ?)
        """, (p["id"], str(CL1), str(V1), cov))

        # ------------------------
        # 2. 7ì¼ê°„ ê¸°ë¡ (2025-09-10 ~ 09-16)
        # ------------------------
        start_date = date.fromisoformat(presc_date1)
        for d in range(days1):
            rec_date = (start_date + timedelta(days=d)).isoformat()
            emo_txt = random.choice(["ì˜¤ëŠ˜ ê´œì°®ìŒ", "ìš°ìš¸í–ˆìŒ", "ë¶ˆë©´ìœ¼ë¡œ í˜ë“¦", "ë¬´ê¸°ë ¥"])
            side_eff = random.choice([0,1,2,3])
            pdc = round(random.uniform(0.6,1.0),2)

            cur.execute("""
                INSERT OR REPLACE INTO patient_daily
                (patient_id, prescription_date, record_date, emotion_text, side_eff_severity, day_num_dose, pdc, medication_id, dose_mg, note)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            """, (p["id"], presc_date1, rec_date, emo_txt, side_eff, day_drug1, pdc, "fluoxetine", dose1, "ì²« ë°©ë¬¸ ê¸°ë¡"))

            phq9_score = random.randint(5, 15)
            cur.execute("""
                INSERT OR REPLACE INTO daily_phq9
                (patient_id, prescription_date, record_date, phq9_score, model_version)
                VALUES (?, ?, ?, ?, ?)
            """, (p["id"], presc_date1, rec_date, phq9_score, "v1.0"))

            if side_eff > 0:
                code = random.choice(["S01","S02"])
                cur.execute("""
                    INSERT OR REPLACE INTO asec_response
                    (patient_id, prescription_date, record_date, side_effect_code, due_to_med, severity)
                    VALUES (?, ?, ?, ?, ?, ?)
                """, (p["id"], presc_date1, rec_date, code, 1, side_eff))

            # daily_predict
            pred_dose = dose1 if d<3 else dose1-5
            cur.execute("""
                INSERT OR REPLACE INTO daily_predict
                (patient_id, predict_dt, pred_dose, pred_frequency)
                VALUES (?, ?, ?, ?)
            """, (p["id"], rec_date, pred_dose, freq1))

        # ------------------------
        # 3. ì¬ë°©ë¬¸ (2025-09-17)
        # ------------------------
        presc_date2 = "2025-09-17"
        dose2 = dose1 + 10 if p["id"]=="test001" else dose1 + 5  # ìš©ëŸ‰ ì¡°ì •
        freq2 = freq1
        days2 = 7
        day_drug2 = int(24/freq2)
        cur.execute("""
            INSERT OR REPLACE INTO patient_predict
            (patient_id, prescription_date, dose, frequency, prescription_days, day_drug, note)
            VALUES (?, ?, ?, ?, ?, ?, ?)
        """, (p["id"], presc_date2, dose2, freq2, days2, day_drug2, "ì¬ë°©ë¬¸ ì²˜ë°© (ìš©ëŸ‰ ì¡°ì •)"))

        # PK param ì—…ë°ì´íŠ¸ (ì¡°ê¸ˆ ë³€í™” ê°€ì •)
        CL2 = round(CL1 * random.uniform(0.9,1.1), 2)
        V2 = round(V1 * random.uniform(0.95,1.05), 2)
        cur.execute("""
            UPDATE pk_param SET pkpram_CL=?, pkpram_V=? WHERE patient_id=?
        """, (str(CL2), str(V2), p["id"]))

        # ------------------------
        # 4. ì¬ë°©ë¬¸ í›„ 7ì¼ ê¸°ë¡ (2025-09-17 ~ 09-23)
        # ------------------------
        start_date2 = date.fromisoformat(presc_date2)
        for d in range(days2):
            rec_date = (start_date2 + timedelta(days=d)).isoformat()
            emo_txt = random.choice(["ì¡°ê¸ˆ ë‚˜ì•„ì§", "ì»¨ë””ì…˜ ì¢‹ìŒ", "ë‘í†µ ìˆìŒ", "ë¶ˆì•ˆ ì§€ì†"])
            side_eff = random.choice([0,1,2])
            pdc = round(random.uniform(0.7,1.0),2)

            cur.execute("""
                INSERT OR REPLACE INTO patient_daily
                (patient_id, prescription_date, record_date, emotion_text, side_eff_severity, day_num_dose, pdc, medication_id, dose_mg, note)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            """, (p["id"], presc_date2, rec_date, emo_txt, side_eff, day_drug2, pdc, "fluoxetine", dose2, "ì¬ë°©ë¬¸ ê¸°ë¡"))

            phq9_score = random.randint(4, 12)
            cur.execute("""
                INSERT OR REPLACE INTO daily_phq9
                (patient_id, prescription_date, record_date, phq9_score, model_version)
                VALUES (?, ?, ?, ?, ?)
            """, (p["id"], presc_date2, rec_date, phq9_score, "v1.1"))

            if side_eff > 0:
                code = random.choice(["S03","S04"])
                cur.execute("""
                    INSERT OR REPLACE INTO asec_response
                    (patient_id, prescription_date, record_date, side_effect_code, due_to_med, severity)
                    VALUES (?, ?, ?, ?, ?, ?)
                """, (p["id"], presc_date2, rec_date, code, 1, side_eff))

            pred_dose = dose2 if d<3 else dose2-5
            cur.execute("""
                INSERT OR REPLACE INTO daily_predict
                (patient_id, predict_dt, pred_dose, pred_frequency)
                VALUES (?, ?, ?, ?)
            """, (p["id"], rec_date, pred_dose, freq2))

    conn.commit()

print("âœ… í™ê¸¸ë™(test001), ì´ëª½ë£¡(test002) â†’ ì²« ì²˜ë°© + 7ì¼ ê¸°ë¡ + ì¬ë°©ë¬¸ PK param ì—…ë°ì´íŠ¸ + ì¬ì²˜ë°© + ê·¸ í›„ 7ì¼ ê¸°ë¡ê¹Œì§€ ì¶”ê°€ ì™„ë£Œ!")
```

```python
import sqlite3
from datetime import datetime
import json

DB_PATH = "project_db2.db"

def add_pk_and_prediction(patient_id, pred_dose, pred_frequency,
                          cl_val, v_val, covariate_dict):
    with sqlite3.connect(DB_PATH) as conn:
        cur = conn.cursor()

        # --- 1) PK parameter ì €ì¥ ---
        cur.execute("""
            INSERT OR REPLACE INTO pk_param (patient_id, pkpram_CL, pkpram_V, covariate)
            VALUES (?, ?, ?, ?)
        """, (patient_id, str(cl_val), str(v_val), json.dumps(covariate_dict, ensure_ascii=False)))

        # --- 2) Daily prediction ì €ì¥ ---
        predict_dt = datetime.now().isoformat(timespec="seconds")
        cur.execute("""
            INSERT INTO daily_predict (patient_id, predict_dt, pred_dose, pred_frequency)
            VALUES (?, ?, ?, ?)
        """, (patient_id, predict_dt, pred_dose, pred_frequency))

        conn.commit()
        print(f"{patient_id} â†’ PK params(CL={cl_val}, V={v_val}), "
              f"Dose={pred_dose}mg, Freq={pred_frequency}h ì €ì¥ ì™„ë£Œ.")

# ---------------------------------
# ì˜ˆì‹œ ì‹¤í–‰
add_pk_and_prediction(
    patient_id="gildong123",
    pred_dose=20,
    pred_frequency=24,
    cl_val=15.3,   # Clearance ê°’ (L/h)
    v_val=120.5,   # Volume ê°’ (L)
    covariate_dict={"WT": 76, "eGFR": 95}
)
```

```python
# In[1]
import sqlite3
import pandas as pd
from datetime import datetime, timedelta

DB_PATH = "project_db2.db"
patient_id = "gildong123"
prescription_date = "2025-09-15"
model_version = "v1.0"

# 7ì¼ê°„ ê¸°ë¡ ë‚ ì§œ ìƒì„±
start_date = datetime(2025, 9, 16)
dates = [(start_date + timedelta(days=i)).date().isoformat() for i in range(7)]

texts = [
    "ì•½ì„ ë¨¹ê¸° ì „ë³´ë‹¤ ì°¨ë¶„í•´ì¡Œì§€ë§Œ ì—¬ì „íˆ ë¬´ê¸°ë ¥í•˜ë‹¤.",
    "ê¸°ë¶„ì´ ì¡°ê¸ˆ ë‚˜ì•„ì¡Œì§€ë§Œ í”¼ê³¤í•˜ë‹¤.",
    "ì˜¤ëŠ˜ì€ ìš°ìš¸ê°ì´ ì‹¬í•˜ì§€ ì•ŠìŒ.",
    "ë¶ˆë©´ìœ¼ë¡œ í”¼ê³¤í•˜ë‹¤.",
    "ë‘í†µì´ ìˆì—ˆì§€ë§Œ ê²¬ë”œë§Œí•˜ë‹¤.",
    "ë¶ˆì•ˆê°ì´ ì¤„ì–´ë“  ë“¯í•˜ë‹¤.",
    "ì „ë°˜ì ìœ¼ë¡œ ì•ˆì •ì ì´ì§€ë§Œ í”¼ê³¤í•¨."
]

side_effects = [("S05", 2), ("S01", 1), ("S02", 2), ("S03", 2), ("S02", 1), ("S04", 1), ("S05", 1)]
pred_doses   = [20, 20, 20, 15, 15, 10, 10]
pred_freqs   = [24, 24, 24, 24, 24, 24, 24]

def estimate_phq9(text: str) -> int:
    neg_words = {"ìš°ìš¸": 3, "ë¬´ê¸°ë ¥": 3, "ë¶ˆë©´": 3, "ë¶ˆì•ˆ": 2, "ë‘í†µ": 2, "í”¼ê³¤": 2, "í”¼ë¡œ": 2, "ìŠ¬í””": 3}
    pos_words = {"ì°¨ë¶„": -1, "ë‚˜ì•„ì¡Œ": -1, "ì•ˆì •": -1}
    score = 0
    for w, val in neg_words.items():
        if w in text: score += val
    for w, val in pos_words.items():
        if w in text: score += val
    return max(0, min(score, 27))

# In[2]
with sqlite3.connect(DB_PATH, timeout=30) as conn:
    # WAL ëª¨ë“œ í™œì„±í™”
    conn.execute("PRAGMA journal_mode=WAL;")
    cur = conn.cursor()

    for i, day in enumerate(dates):
        phq9_score = estimate_phq9(texts[i])

        # patient_daily
        cur.execute("""
            INSERT INTO patient_daily 
            (patient_id, prescription_date, record_date, emotion_text, side_eff_severity, day_num_dose, pdc, medication_id, dose_mg, note)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (patient_id, prescription_date, day, texts[i], side_effects[i][1], 1, 0.95, "fluoxetine", pred_doses[i], "ìë™ ì…ë ¥"))

        # asec_response
        cur.execute("""
            INSERT INTO asec_response 
            (patient_id, prescription_date, record_date, side_effect_code, due_to_med, severity)
            VALUES (?, ?, ?, ?, ?, ?)
        """, (patient_id, prescription_date, day, side_effects[i][0], 1, side_effects[i][1]))

        # daily_predict
        cur.execute("""
            INSERT INTO daily_predict 
            (patient_id, predict_dt, pred_dose, pred_frequency)
            VALUES (?, ?, ?, ?)
        """, (patient_id, day, pred_doses[i], pred_freqs[i]))

        # daily_phq9 (upsert)
        cur.execute("""
            INSERT INTO daily_phq9 (patient_id, prescription_date, record_date, phq9_score, model_version)
            VALUES (?, ?, ?, ?, ?)
            ON CONFLICT(patient_id, prescription_date, record_date)
            DO UPDATE SET
                phq9_score=excluded.phq9_score,
                model_version=excluded.model_version
        """, (patient_id, prescription_date, day, phq9_score, model_version))

    conn.commit()

print("âœ… 7ì¼ì¹˜ ê¸°ë¡ì´ WAL ëª¨ë“œì—ì„œ ì•ˆì „í•˜ê²Œ ì¶”ê°€/ì—…ë°ì´íŠ¸ ë˜ì—ˆìŠµë‹ˆë‹¤.")

# In[3]
# í™•ì¸
with sqlite3.connect(DB_PATH) as conn:
    df = pd.read_sql_query(f"""
        SELECT record_date, phq9_score, model_version
        FROM daily_phq9
        WHERE patient_id='{patient_id}'
        ORDER BY record_date
    """, conn)
display(df)
```
